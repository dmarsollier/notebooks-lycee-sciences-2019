# Tracer un histogramme de donnnées

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

# Gaussienne

```python
x = np.linspace(0, 20, 100)
```

```python
# moyenne
m = 11.5
# écart-type
s = 1.2
```

```python
y = 1/(s*math.sqrt(2*math.pi)) * np.exp(-(x-m)**2 / (2*s**2))
```

```python
fig, ax = plt.subplots()
plt.plot(x, y)
```

# Gaussienne en histogramme

```python
y = [0]*8 + [3, 14, 41, 80, 100, 41, 14, 3] + [0]*4
```

```python
x = np.arange(20)
```

```python
fig, ax = plt.subplots()
plt.bar(x, y, align='edge')
plt.show()
```
